```python
import os
import shutil
```

```python
def move_txt_files(source_folder, destination_folder):
    """
    Переносит все файлы с расширением .txt из source_folder в destination_folder.

    :param source_folder: Путь к исходной папке с файлами
    :param destination_folder: Путь к папке, куда нужно перенести файлы
    """
    # Проверяем, существует ли исходная папка
    if not os.path.exists(source_folder):
        print(f"Исходная папка не существует: {source_folder}")
        return

    # Создаем целевую папку, если её нет
    os.makedirs(destination_folder, exist_ok=True)

    # Получаем список всех файлов в исходной папке
    files = os.listdir(source_folder)

    # Фильтруем только .txt файлы
    txt_files = [file for file in files if file.endswith('.jpg')]

    # Переносим каждый .txt файл
    for txt_file in txt_files:
        source_path = os.path.join(source_folder, txt_file)
        destination_path = os.path.join(destination_folder, txt_file)

        try:
            shutil.move(source_path, destination_path)
            print(f"Файл {txt_file} успешно перенесен.")
        except Exception as e:
            print(f"Ошибка при переносе файла {txt_file}: {e}")


move_txt_files("/content/drive/MyDrive/Диплом/dataset/Original Dataset", "/content/drive/MyDrive/Диплом/dataset/Original Dataset/images")
```
